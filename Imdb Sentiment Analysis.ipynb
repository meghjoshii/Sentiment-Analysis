# Data Cleansing

```python
# lowercase, remove html elements, punctuations, stop words, n-grams
```

```python
import re
```

```python
remove_html_elements=re.compile("(<br\s*/><br\s*/>)|(\/)")
```

```python
remove_punctuations=re.compile("(\.)|(\;)|(\:)|(\!)|(\')|(\?)|(\,)|(\")|(\()|(\))|(\[)|(\])|(\d+)")
```

```python
reviews_train1= []
```

```python
reviews_train2= []
```

```python
reviews_test1=[]
```

```python
reviews_test2=[]
```

```python
reviews_test3=[]
```

```python
#removing punctuations, html elements and converting to lower case
def preprocess(reviews):
    for line in reviews:
        reviews_train1.append(line.lower())  
    for line in reviews_train1:
        reviews_train2.append(remove_html_elements.sub(" ",str(line).lower()))
    for line in reviews_train2:
        reviews_train3.append(remove_punctuations.sub(" ",str(line).lower()))
    return reviews_train3
```

```python
def remove_sw(corpus):
    removed_sw=[]
    for review in corpus:
        removed_sw.append(
            ' '.join([word for word in review.split() 
                      if word not in english_stop_words])
        )
    return removed_sw
    
    
```

```python
#stemming - stemming is the process of reducing inflected (or sometimes derived) words to their word stem, base or root form
#Porter & Lancaster - porter is faster
```

```python
from nltk.stem.porter import PorterStemmer
```

```python
def stem(corpus):
    stemmer = PorterStemmer()
    return [' '.join([stemmer.stem(word) for word in review.split()]) for review in corpus]
```

```python
def get_lemmatized(corpus):
    lemmatizer = WordNetLemmatizer()
    return [' '.join([lemmatizer.lemmatize(word) for word in review.split()]) for review in corpus]
```
